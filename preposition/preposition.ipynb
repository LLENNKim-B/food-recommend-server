```python
import pandas as pd

df = pd.read_csv('disease.csv')
```

```python
df.head(10)
```

```python
for idx, row in df.iterrows():
    print(f"Disease: {row['disease']}")
    print(f"  Avoid Nutrients: {row['avoid_nutrients']}")
    print(f"  Avoid Ingredients: {row['avoid_ingredients']}")
    print(f"  Note: {row['note']}\n")
```

```python
df.describe()
```

```python
df.info()
```

```python
no_columns = ['avoid_nutrients', 'avoid_ingredients']
df[df[no_columns] == "no"].count()
```

```python
df[no_columns] = df[no_columns].replace("no",'none')
```

```python
df.head(20)
```

```python
df.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# สร้าง TF-IDF จากคอลัมน์ note
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['note'].fillna(""))

# แปลงเป็น DataFrame เพื่อดูค่า TF-IDF
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df.head(10)
```

```python
df = pd.concat([df.reset_index(drop=True), tfidf_df.reset_index(drop=True)], axis=1)
```

```python
df.head()
```

```python
# บันทึก df_final เป็นไฟล์ CSV
df.to_csv("disease_prep.csv", index=False, encoding="utf-8-sig")
```
